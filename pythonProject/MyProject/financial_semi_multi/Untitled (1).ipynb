```python
data = pd.read_csv('./../../data/inf_uem_btc_date.csv')
data
```

```python
itr = pd.read_csv('./../../data/FEDFUNDS.csv')
itr['DATE'] = pd.to_datetime(itr['DATE'])
fed = pd.DataFrame()
for year in range(2010, 2023):
    fed = fed.append(itr[itr['DATE'] == f'{year}-04-01'], ignore_index=True)
    fed = fed.append(itr[itr['DATE'] == f'{year}-07-01'], ignore_index=True)
    fed = fed.append(itr[itr['DATE'] == f'{year}-10-01'], ignore_index=True)
    fed = fed.append(itr[itr['DATE'] == f'{year+1}-01-01'], ignore_index=True)
fed = fed.drop(index=0, axis = 0)
fed
```

```python
btc = pd.read_csv('./../../data/investing_crowl.csv')
btc['Date'] = pd.to_datetime(btc['Date'])
bt = pd.DataFrame()
for year in range(2010, 2023):
    bt = bt.append(btc[btc['Date'] == f'{year}-03-31'], ignore_index=True)
    bt = bt.append(btc[btc['Date'] == f'{year}-06-30'], ignore_index=True)
    bt = bt.append(btc[btc['Date'] == f'{year}-09-30'], ignore_index=True)
    bt = bt.append(btc[btc['Date'] == f'{year}-12-31'], ignore_index=True)
bt = bt.drop(index=0, axis = 0)
bt
```
