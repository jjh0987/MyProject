```python
data_s = [pd.read_csv(i) for i in data_path[5:]]
# tar = data_s[0]
# for i in tar.columns:
#     plt.title(i)
#     tar[i].plot()
#     plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from tqdm import tqdm

c_index = ['c30','c40','c50','c70','c100']
s_index = ['s30','s40','s50','s70','s100']

answer = pd.DataFrame()

for j in tqdm(range(len(data_s))):
    tar = data_s[j]
    train = tar[:8001]
    test = tar[10001:]

    # train_x = train.iloc[:8001,:30]
    # valid_x = train.iloc[8001:,:30]
    train_x = train.iloc[:,:35]
    test_x = test.iloc[:,:35]
    for i in range(4):
        print(f'{target[i]}_{s_index[j]}')
    #     train_y = train.loc[:8000,target[i]]
    #     valid_y = train.loc[8001:,target[i]]
        train_y = train.loc[:,target[i]]

        cat = RandomForestRegressor(random_state=77) # CatBoostRegressor(random_state=77,iterations=2000,learning_rate=0.001,loss_function='MAPE',verbose=1000)
        cat.fit(train_x,train_y)
        pred = cat.predict(test_x)

        answer[f'{target[i]}_{s_index[j]}'] = pred
        # print(1.2*mean_absolute_error(pred,valid_y))
    print('-'*70)

for j in tqdm(range(len(data_c))):
    tar = data_c[j]
    train = tar[:8001]
    test = tar[10001:]

    # train_x = train.iloc[:8001,:30]
    # valid_x = train.iloc[8001:,:30]
    train_x = train.iloc[:,:33]
    test_x = test.iloc[:,:33]
    for i in range(4):
        print(f'{target[i]}_{c_index[j]}')
    #     train_y = train.loc[:8000,target[i]]
    #     valid_y = train.loc[8001:,target[i]]
        train_y = train.loc[:,target[i]]

        cat = RandomForestRegressor(random_state=77) # CatBoostRegressor(random_state=77,iterations=2000,learning_rate=0.001,loss_function='MAPE',verbose=1000)
        cat.fit(train_x,train_y)
        pred = cat.predict(test_x)

        answer[f'{target[i]}_{c_index[j]}'] = pred
        # print(1.2*mean_absolute_error(pred,valid_y))
    print('-'*70)
    
```
