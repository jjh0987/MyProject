```python
def rnn_data(data,label_name,date_range=1):
    if label_name:
        data = data.reset_index(drop=True)
        y_train = []
        for i in range(date_range-1,len(data)):
            y_train.append(data.loc[i,label_name])
        data = data.drop(label_name,axis=1)
        x_train = []
        for i in range(date_range-1,len(data)):
            x_train.append(np.array(data.loc[i-(date_range-1):i,:]))
        return np.array(x_train), np.array(y_train)
    else:
        data = data.reset_index(drop=True)
#         y_train = []
#         for i in range(date_range-1,len(data)):
#             y_train.append(data.loc[i,label_name])
#         data = data.drop(label_name,axis=1)
        x_train = []
        for i in range(date_range-1,len(data)):
            x_train.append(np.array(data.loc[i-(date_range-1):i,:]))
        return np.array(x_train)
    
def LSTM(x_train,y_train,
         activation,epochs,batch_size,lr,
         loss='mean_squared_error',
         drop_rate=0.1):
    
    def scheduler(epoch, lr):
        if epoch < 5:
            return lr
        else:
            return lr * tf.math.exp(-0.1)
    
    Sequential = tf.keras.models.Sequential
    Dense = tf.keras.layers.Dense
    LSTM = tf.keras.layers.LSTM
    DROP = tf.keras.layers.Dropout
    # keras.layers.GRU()
    # if scaler:
    #     scaler = scaler()
    #     scaler.fit_transform(target_columns)
    model = Sequential()
    model.add(LSTM(
        8,
        input_shape=x_train[0].shape,
        activation=activation
    ))
    model.add(DROP(drop_rate))
    model.add(Dense(1))


    optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
    model.compile(loss=loss,optimizer=optimizer)
    
    callback = tf.keras.callbacks.LearningRateScheduler(scheduler)

    model.fit(x_train,y_train,epochs=epochs,batch_size=batch_size)

    return model

answer = []
spliter = 68
for i in tqdm(range(len(train_c))):
    x_train,y_train = rnn_data(train_c[i].loc[:10000,target_],target_[3],spliter)
    model = LSTM(x_train,y_train,'relu',epochs=10,batch_size=4,lr=0.005,loss='mape')
    tar_test = pd.concat([train_c[i].drop([target_[3]],axis=1)[10000-spliter:],test_c[i]]).reset_index(drop=True)
    test = rnn_data(tar_test,False,spliter)
    pred = model.predict(test)
    answer.extend(pred)
```
