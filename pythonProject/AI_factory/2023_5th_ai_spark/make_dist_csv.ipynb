```python
import pandas as pd
aws = pd.read_csv('./META/awsmap.csv')
pm = pd.read_csv('./META/pmmap.csv')
aws = aws.sort_values('Location').reset_index(drop=True)
pm = pm.sort_values('Location').reset_index(drop=True)
```

```python
pm
```

```python
cols = pm['Location']
index = aws['Location']
```

```python
from haversine import haversine, Unit, haversine_vector
start = [(i-90,j) for i,j in zip(pm['Longitude'],pm['Latitude'])]
end = [(i-90,j) for i,j in zip(aws['Longitude'],aws['Latitude'])]

distance = []
for i in end:
    tp = []
    for j in start:
        tp.append(haversine(i,j, unit=Unit.KILOMETERS ))
    distance.append(tp)
distance = pd.DataFrame(distance,columns=cols)
distance.index = index
distance
```
