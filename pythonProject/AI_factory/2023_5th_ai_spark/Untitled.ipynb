```python
def find_range(target,center_range,loc_type='train'):
    answer = []
    for i in dt_tbl[dt_tbl[target] < center_range].index:
        if loc_type == 'train':
            answer.append(pd.read_csv(f'./TRAIN_AWS/{i}.csv'))
        else:
            answer.append(pd.read_csv(f'./TEST_AWS/{i}.csv'))
    return answer

target = '공주'
label = pd.read_csv(f'./TRAIN/{target}.csv')
tar = find_range(target,200,loc_type='train')
test_tar = find_range(target,200,loc_type='test')
# [i[i['연도'] == 3] for i in tar]
```

```python
mean_col = ['기온(°C)','풍향(deg)','풍속(m/s)','습도(%)']
answer = []

for target in dt_tbl.index:
    test_tar = find_range(target,200,loc_type='test')

    test_df = pd.DataFrame()
    test_df['연도'] = test_tar[0]['연도']
    test_df['일시'] = test_tar[0]['일시']
    # new_df['지점'] = [target for _ in range(len(tar[0]))]

    for idx in mean_col:
        answer = []
        for i in test_tar[0].index:
            tp = pd.Series([pd.NA])
            cnt = 1
            for tbl in tar:
                tp[cnt] = tbl.loc[i,idx]
                cnt += 1
            answer.append(tp.median())
        test_df[idx] = answer

    test_df['월'] = test_df['일시'].str[:2].astype(int)/12.
    test_df['일'] = test_df['일시'].str[3:5].astype(int)/31.
    test_df['시'] = test_df['일시'].str[6:8].astype(int)/23.
    test_df = test_df.drop(['일시','연도'],axis=1)
    answer.append()
```
