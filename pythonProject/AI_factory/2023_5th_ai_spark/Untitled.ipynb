```python
def find_range(target,center_range):
    answer = []
    for i in dt_tbl[dt_tbl[target] < center_range].index:
        answer.append(pd.read_csv(f'./TRAIN_AWS/{i}.csv'))
    return answer

target = '공주'
tar = find_range(target,20)
tar 
# [i[i['연도'] == 3] for i in tar]
```

```python
from catboost import CatBoostRegressor
```

```python
mean_col = ['기온(°C)','풍향(deg)','풍속(m/s)','습도(%)']
new_df = pd.DataFrame()
new_df['연도'] = tar[0]['연도']
new_df['일시'] = tar[0]['일시']
# new_df['지점'] = [target for _ in range(len(tar[0]))]

for idx in mean_col:
    answer = []
    for i in tar[0].index:
        tp = pd.Series([pd.NA])
        cnt = 1
        for tbl in tar:
            tp[cnt] = tbl.loc[i,idx]
            cnt += 1
        answer.append(tp.median())
    new_df[idx] = answer

new_df['월'] = new_df['일시'].str[:2].astype(int)/12.
new_df['일'] = new_df['일시'].str[3:5].astype(int)/31.
new_df['시'] = new_df['일시'].str[6:8].astype(int)/23.
new_df = new_df.drop('일시',axis=1)
label = pd.read_csv(f'./TRAIN/{target}.csv')
```

```python
new_df
```

```python
new_df = new_df.dropna()
label = label.loc[new_df.index,:]
new_df = new_df.drop('연도',axis=1)
```

```python
cat = CatBoostRegressor(random_seed=11)
cat.fit(new_df,label['PM2.5'])
```
