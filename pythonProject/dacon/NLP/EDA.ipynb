```python
from sklearn.model_selection import StratifiedShuffleSplit
train = train[train['Speaker'] == 'Joey']
train = train.reset_index().drop('index',axis=1)
y = train.Target
X = train.drop('Target',axis=1)
skf = StratifiedShuffleSplit(n_splits=2, test_size=0.1, random_state=0)
skf.get_n_splits(X, y)

train_index, test_index = skf.split(X, y)

print(train.loc[train_index[0],:].Target.value_counts())
print(train.loc[train_index[1],:].Target.value_counts())
```
