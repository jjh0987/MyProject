```python
train = pd.read_csv('../train.csv')
```

```python
target = ['SNP_01', 'SNP_02',
       'SNP_03', 'SNP_04', 'SNP_05', 'SNP_06', 'SNP_07', 'SNP_08', 'SNP_09',
       'SNP_10', 'SNP_11', 'SNP_12', 'SNP_13', 'SNP_14', 'SNP_15']
ans = []
for i in range(len(train)):
    tp = ''
    for j in target:
        tp += train.loc[i,j]+' '
    ans.append(tp)
```

```python
x = pd.DataFrame(ans)
x['class'] = train['class']
x
```

```python
A = x[x['class']=='A'].drop_duplicates()
A
```

```python
B = x[x['class']=='B'].drop_duplicates()
B
```

```python
C= x[x['class']=='C'].drop_duplicates()
C
```

```python
cnt = 0
for i in list(A.loc[:,0]):
    if i in list(C.loc[:,0]):
        cnt += 1
cnt
```

```python
test = pd.read_csv('../test.csv')
target = ['SNP_01', 'SNP_02',
       'SNP_03', 'SNP_04', 'SNP_05', 'SNP_06', 'SNP_07', 'SNP_08', 'SNP_09',
       'SNP_10', 'SNP_11', 'SNP_12', 'SNP_13', 'SNP_14', 'SNP_15']
ans = []
for i in range(len(test)):
    tp = ''
    for j in target:
        tp += test.loc[i,j]+' '
    ans.append(tp)
y = pd.DataFrame(ans)
test
```

```python
cnt = 0
erase = []
for i in list(y.loc[:,0]):
    if i in list(A.loc[:,0]):
        erase.append(i)
    elif i in list(B.loc[:,0]):
        erase.append(i)
    elif i in list(C.loc[:,0]):
        erase.append(i)
erase
```

```python
train
```

```python
for i in range(len(ans)):
    if ans[i] in erase:
        ans[i] = ''
train['concat'] = ans
train
```

```python
train[train['concat'] != ''].drop('concat',axis=1).to_csv('./erase_target.csv')
```

```python
snp_col = [f'SNP_{str(x).zfill(2)}' for x in range(1,16)]
tp = []
for col in snp_col:
    tp.append(list(train[col].value_counts().index))
temp = []
for i in tp:
    temp.extend(i)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(sorted(list(set(temp))))
for col in snp_col:
    train[col] = le.transform(train[col])
```
