```python
import pandas as pd

train = pd.read_csv('./train.csv')
```

```python
train = train.drop('종목명',axis=1)
train = train.rename(columns={'일자':'date','종목코드':'code','거래량':'volume','시가':'start','고가':'high',
                             '저가':'low','종가':'close'})
train = train.sort_values(['code','date'])
train = train.reset_index(drop=True)
```

```python
code = train['code'].unique()
code.sort()
code
```

```python
train
```

```python
initial = list(train[train['date']==20230530]['close']) 
target = list(train.groupby('code').mean()['close'])

ans = []
for i in range(len(initial)):
    ans.append((initial[i] - target[i])/target[i])
```

```python
ans
```

```python
sub = pd.read_csv('./sample_submission.csv')
sub['ans'] = ans
sub = sub.drop('순위',axis=1)
sub = sub.sort_values('ans',ascending=False)
sub['순위'] = [i for i in range(1,2001)]
sub = sub.drop('ans',axis=1)
sub = sub.sort_values('종목코드')
sub.to_csv('./submission.csv')
```
