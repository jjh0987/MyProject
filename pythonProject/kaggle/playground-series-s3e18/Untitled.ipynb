```python
train = pd.read_csv('./train.csv')
train.drop(['EC1','EC2','EC4','EC6','EC5','id'],axis=1,inplace=True)
train.head()
```

```python
from pycaret import classification
classification.setup(data=train,
                     target='EC3',
                     fold_strategy='kfold',
                     train_size=0.8,
                     session_id=77,
                     remove_outliers=True,
                     outliers_threshold=0.02)
                     # outliers_method='iforest')
```
