```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
from sklearn.metrics import roc_auc_score
from pycaret import classification
from tqdm import tqdm
from sklearn.preprocessing import MinMaxScaler

warnings.filterwarnings("ignore")
```

```python
cols = ['BertzCT', 'Chi1', 'Chi1n', 'Chi1v', 'Chi2n', 'Chi2v', 'Chi3v', 'Chi4n',
       'EState_VSA1', 'EState_VSA2', 'ExactMolWt', 'FpDensityMorgan1',
       'FpDensityMorgan2', 'FpDensityMorgan3', 'HallKierAlpha',
       'HeavyAtomMolWt', 'Kappa3', 'MaxAbsEStateIndex', 'MinEStateIndex',
       'NumHeteroatoms', 'PEOE_VSA10', 'PEOE_VSA14', 'PEOE_VSA6', 'PEOE_VSA7',
       'PEOE_VSA8', 'SMR_VSA10', 'SMR_VSA5', 'SlogP_VSA3', 'VSA_EState9',
       'fr_COO', 'fr_COO2']

# train = pd.read_csv('./train.csv')
# train.drop(['id'],axis=1,inplace=True)
train = pd.read_csv('./train.csv')
train.drop(['id'],axis=1,inplace=True)
# train.head()
```

```python
# train.drop(['id','EC1','EC2','EC4','EC5','EC6'],axis=1,inplace=True)
target = ['EC3','EC4','EC5','EC6']

test = pd.read_csv('./test.csv')
test = test.drop('id',axis=1)

# tar = target.pop()
# cols.append(tar)
tar = 'EC1'
cols.append(tar)
classification.setup(data=train.loc[:,cols],
                     target=tar,
                     fold_strategy='kfold',
                     train_size=0.8,
                     session_id=77)
                     # remove_outliers=True)
                     # outliers_threshold=0.02)
                     # outliers_method='iforest')

best = classification.compare_models(n_select=16,fold=10,sort='AUC')
j = 0
k = 0
pred = np.array([0 for _ in range(len(test))]).astype('float')
while j < 5:
    try:
        pred += np.array([i[1] for i in best[j].predict_proba(test)])
        j += 1
        k += 1
    except:
        j += 1
        pass
pred = [i/k for i in pred]
print(pred[:10])
# pred = np.array([best[i].predict_proba(test) for i in range(3)]).mean(0)
test[tar] = pred
```
